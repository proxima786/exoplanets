```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#RATIO 1- Ds/Dp
#----------------------------------------------------------------------------------------------------------------------------------
df=pd.read_csv('1st plot ratio-1.csv')
plt.figure(figsize=(10, 6))
plt.scatter(df['pl_rade (Dp)'],df['Ds/Dp'], color='blue', alpha=0.7)
plt.xlabel('Diameter of planet (Dp)')
plt.ylabel('Ds/Dp (Stellar to planetary diameter)')
plt.title('Plot of Ds/Dp vs Dp')
plt.grid(True)
plt.show()
```

```python
#Ruling out exoplanets which are out of the range.
#Range taken- Diameter should lie between 0.5-2 times earth diameter
# Ratio should lie between between 80 and 125
#------------------------------------------------------------------------------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('1st plot ratio-1.csv')
plt.figure(figsize=(10, 6))
plt.scatter(df['pl_rade (Dp)'], df['Ds/Dp'], color='blue', alpha=0.7)
x_lower_limit = 0.5
x_upper_limit = 2.0
y_lower_limit = 80
y_upper_limit = 125
filtered_data = df[
    (df['pl_rade (Dp)'] >= x_lower_limit) & (df['pl_rade (Dp)'] <= x_upper_limit) &
    (df['Ds/Dp'] >= y_lower_limit) & (df['Ds/Dp'] <= y_upper_limit)]
filtered_df = pd.DataFrame(filtered_data)
plt.scatter(filtered_df['pl_rade (Dp)'], filtered_df['Ds/Dp'], color='red', alpha=0.7)
plt.xlabel('Diameter of planet (Dp)')
plt.ylabel('Ds/Dp (Stellar to planetary diameter)')
plt.title('Plot of Ds/Dp vs Dp')
plt.grid(True)
plt.show()
print('Exoplanets in the given range:')
filtered_df
```
