```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('Exoplanet-file1.csv')

# Convert 'Ds/Dp' column to numeric values, ignoring errors
df['Ds/Dp'] = pd.to_numeric(df['Ds/Dp'], errors='coerce')

# Filter for rows with 'Ds/Dp' values between 75 and 130
filtered_df = df[(df['Ds/Dp'] >= 75) & (df['Ds/Dp'] <= 130)]

# Group by exoplanet name and keep the row with the maximum 'Ds/Dp' value within the range
grouped = filtered_df.groupby('Exoplanet')
result_df = grouped.apply(lambda x: x.loc[x['Ds/Dp'].idxmax()])

# Reset index to remove the multi-level index added by the groupby operation
result_df.reset_index(drop=True, inplace=True)
import pandas as pd

# Assume you have a DataFrame named df and the column to drop is 'Diameter of planet in terms of Earth\'s diameter'
column_to_drop = "Diameter of planet in terms of Earth's diameter"
column_to_drop2 = "Diameter of star in terms of Sun's diameter"
result_df.drop(columns=[column_to_drop], inplace=True)

result_df.drop(columns=[column_to_drop2], inplace=True)


# Print the resulting DataFrame
result_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(result_df['Dp'], result_df['Ds/Dp'], color='blue', alpha=0.7)
plt.xlabel('Ds/Dp')
plt.ylabel('Dp')
plt.title('Plot of Ds/Dp vs Dp')



plt.grid(True)
plt.show()





```

```python
print('before filter-1:',len(df))
print('after-filter-1:',len(result_df))
```

```python
df['d/Ds']=pd.to_numeric(df['d/Ds'],errors='coerce')
x=df['d/Ds']>=75
y=df['d/Ds']<=130
z= df[(x) & (y)]
z
grouped = z.groupby('Exoplanet')
z1 = grouped.apply(lambda x: x.loc[x['d/Ds'].idxmax()])
column_to_drop = "Diameter of planet in terms of Earth's diameter"
column_to_drop2 = "Diameter of star in terms of Sun's diameter"
z1.drop(columns=[column_to_drop], inplace=True)

z1.drop(columns=[column_to_drop2], inplace=True)

# Reset index to remove the multi-level index added by the groupby operation
z1.reset_index(drop=True, inplace=True)
z1
```
