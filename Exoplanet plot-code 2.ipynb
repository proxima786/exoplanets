```python
from matplotlib import rc
import matplotlib.pyplot as plt
import pandas as pd

# Set the figure size using rc
rc('figure', figsize=(25, 10))

# Read the CSV file into a DataFrame
df = pd.read_csv('1st plot ratio-1.csv')

# Create a scatter plot
plt.figure()
for index, row in df.iterrows():
    color = 'red' if row['pl_name (Exoplanet)'] == 'Earth' else 'blue'
    plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color=color, alpha=0.7)

plt.xlabel('Diameter of planet (Dp)')
plt.ylabel('Diameter of star (Ds)')
plt.title('Plot of Ds v/s Dp')

# Invert the x-axis
plt.gca().invert_xaxis()

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file into a DataFrame
dff = pd.read_csv('2nd plot ratio-2.csv')

# Create a scatter plot
plt.figure(figsize=(10, 6))
for index, row in dff.iterrows():
    color = 'red' if row['pl_name (Exoplanet)'] == 'Earth' else 'blue'
    plt.scatter(row['st_rad (Ds)'], row['pl_orbsmax (d)'], color=color, alpha=0.7)

plt.xlabel('Ds')
plt.ylabel('d')
plt.title('Plot of Ds v/s d')

# Invert the x-axis
plt.gca().invert_xaxis()

plt.grid(True)
plt.show()
```
