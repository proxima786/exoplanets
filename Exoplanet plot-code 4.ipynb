```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Reading 

```python
df=pd.read_csv('1st plot ratio-1.csv')
df.drop(['Ds/Dp','d/Ds'],axis=1)
```

```python
kepler_452b_df = df[df['pl_name (Exoplanet)'] == 'Kepler-62 f']

# Print the resulting DataFrame
kepler_452b_df
```

```python
v= input(print("exoplanet?"))

if v in df['pl_name (Exoplanet)'].values:
    print(f"{v} exists in the 'pl_name' column.")
else:
    print(f"{v} does not exist in the 'pl_name' column.")
```

```python
earth = ['Earth', 'sun', 1,1, 1, 1, 1]  # Replace 0s with your data
koi= ['KOI-4878.01', 'KOI-4878',1.137,1.04,1.131,1.04]
lhs=['LHS 1140 b','LHS 1140',0.0957,1.635,1.730,1.635,1.730,116.126,5.957]
df.loc[len(df)] = earth
df.loc[len(df)]= koi
df.loc[len(df)]= lhs
```

```python
if 'LHS1 1140b' in df['pl_name (Exoplanet)'].values:
    print("LHS-1140 b exists in the 'pl_name' column.")
else:
    print("LHS-1140 b does not exist in the 'pl_name' column.")
```

# Plotting

```python
from matplotlib.lines import Line2D

plt.figure(figsize=(10, 6))

# Iterate through the DataFrame and plot scatter points
for index, row in df.iterrows():
    color = 'yellow' if row['pl_name (Exoplanet)'] == 'Earth' else 'blue'
    plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color=color, alpha=0.7)

    # Highlight Kepler-452 b and Kepler-62f with special markers
    if row['pl_name (Exoplanet)'] == 'Kepler-452 b':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='red', marker='*', s=100, label='Kepler-452 b')
    if row['pl_name (Exoplanet)'] == 'Kepler-62 f':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='purple', marker='*', s=100, label='Kepler-62 f')
    
    if row['pl_name (Exoplanet)']== 'KOI-4878.01':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='black', marker='*', s=100, label='KOI-4878.01')
    if row['pl_name (Exoplanet)']== 'LHS 1140 b':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='green', marker='*', s=100, label='KOI-4878.01')
        
# Plot the line y = x passing through the origin
max_value = max(df['pl_rade (Dp)'].max(), df['st_rad (Ds)'].max())
plt.plot([0, max_value], [0, max_value], color='green', linestyle='dashed', label='y = x')

# Create custom legend entries for different markers
legend_entry = Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=8, label='Exoplanets')
legend_entry_1 = Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=8, label='Earth')

legend_entry_3 = Line2D([0], [0], marker='*', color='red', markersize=10, label='Kepler-452 b')
legend_entry_4 = Line2D([0], [0], marker='*', color='purple', markersize=10, label='Kepler-62 f')
legend_entry_5 = Line2D([0], [0], marker='*', color='black', markersize=10, label='KOI-4878.01')
legend_entry_6 = Line2D([0], [0], marker='*', color='green', markersize=10, label='LHS 1140 b')


# Set the x-axis and y-axis limits
plt.xlim(0, 2)
plt.ylim(0, 2)

# Add the custom legend entries along with the line legend
plt.legend(handles=[legend_entry_1, legend_entry,legend_entry_3, legend_entry_4,legend_entry_5,legend_entry_6], loc='upper left')

# Use LaTeX-style formatting for the axis labels
plt.xlabel('Diameter of planet ($D_p$)')
plt.ylabel('Diameter of star ($D_s$)')
plt.title('Plot of $D_s$ $v/s$ $D_p$')
plt.grid(True)
plt.savefig('exoplanet_plot.pdf', format='pdf', bbox_inches='tight')
plt.show()
dff=pd.read_csv('2nd plot ratio-2.csv')
dff

# Create a list with your own data for the new row
earth = ['Earth', 'sun', 1,1, 1, 1, 1,109.7522, 107.1428]  # Replace 0s with your data
koi= ['KOI-4878.01', 'KOI-4878',1.137,1.04,1.131,1.04,1.131,138.24,108.275]
lhs=['LHS 1140 b','LHS 1140',0.0957,1.635,1.730,1.635,1.730,116.126,5.957]


# Add the new row to the DataFrame
dff.loc[len(dff)] = earth
dff.loc[len(dff)]= koi
dff.loc[len(dff)]= lhs


# Reset the index
dff = dff.reset_index(drop=True)

# Print the DataFrame after adding the new row
dff

plt.figure(figsize=(10, 6))
for index, row in dff.iterrows():
    color = 'yellow' if row['pl_name (Exoplanet)'] == 'Earth' else 'blue'
    plt.scatter(row['st_rad (Ds)'], row['pl_orbsmax (d)'], color=color, alpha=0.7)
    
    if row['pl_name (Exoplanet)'] == 'Kepler-452 b':
        plt.scatter(row['st_rad (Ds)'], row['pl_orbsmax (d)'], color='red', marker='*', s=100, label='Kepler-452 b')
        
    if row['pl_name (Exoplanet)'] == 'Kepler-62 f':
        plt.scatter(row['st_rad (Ds)'], row['pl_orbsmax (d)'], color='purple', marker='*', s=100, label='Kepler-62 f')
        
    if row['pl_name (Exoplanet)']== 'KOI-4878.01':
        plt.scatter(row['st_rad (Ds)'], row['pl_orbsmax (d)'], color='black', marker='*', s=100, label='KOI-4878.01')
    if row['pl_name (Exoplanet)']== 'LHS 1140 b':
        plt.scatter(row['st_rad (Ds)'], row['pl_orbsmax (d)'], color='green', marker='*', s=100, label='LHS 1140 b')

# Plot the line y = x passing through the origin
max_value = max(df['st_rad (Ds)'].max(), df['pl_orbsmax (d)'].max())
plt.plot([0, 6], [0, 6], color='green', linestyle='dashed')

# Create a custom legend entry for the blue data points
legend_entry = Line2D([0], [0], marker='*', color='w', markerfacecolor='blue', markersize=8, label='Exoplanets')
legend_entry_1 = Line2D([0], [0], marker='*', color='w', markerfacecolor='yellow', markersize=8, label='Earth')

legend_entry_3 = Line2D([0], [0], marker='*', color='red', markersize=10, label='Kepler-452 b')
legend_entry_4 = Line2D([0], [0], marker='*', color='purple', markersize=10, label='Kepler-62 f')
legend_entry_5 = Line2D([0], [0], marker='*', color='black', markersize=10, label='KOI-4878.01')
legend_entry_6 = Line2D([0], [0], marker='*', color='green', markersize=10, label='LHS 1140 b')


# Set the x-axis and y-axis limits
plt.xlim(0, 2)
plt.ylim(0, 2)

# Add the custom legend entry along with the line legend
plt.legend(handles=[legend_entry_1, legend_entry,legend_entry_3,legend_entry_4,legend_entry_5,legend_entry_6], loc='upper left')

plt.xlabel('$D_s$(Diameter of Star)')
plt.ylabel('d(Semi-major axis in AU)')
plt.title('Plot of d $v/s$ $D_s$')
plt.grid(True)
plt.savefig('exoplanet_plot2.pdf',format='pdf', bbox_inches='tight')

plt.show()
```
