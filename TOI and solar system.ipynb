```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('1st plot ratio-1.csv')
dff=df.drop(['Ds/Dp','d/Ds'],axis=1)
dff
```

```python
kepler_452b_df = dff[dff['pl_name (Exoplanet)'] == 'TOI-1136 c']

# Print the resulting DataFrame
kepler_452b_df
```

```python
earth = ['Earth', 'sun', 1, 1, 1,1,1]
koi = ['KOI-4878.01', 'KOI-4878', 1.137, 1.04, 1.131,1,1]
lhs = ['LHS 1140 b', 'LHS 1140', 0.0957, 1.635, 1.730,1,1]
toib=['TOI-1136 b','TOI-1136',0.05106,1.9,0.968,1.9,0.968]
toic=['TOI-1136 c','TOI-1136',0.0669,0.7175,0.968,0.7175,0.968]
toid=['TOI-1136 d','TOI-1136',0.1062,1.184,0.968,1.184,0.968]
toie=['TOI-1136 e','TOI-1136',0.139,0.67445,0.968,0.67445,0.968]
toif=['TOI-1136 f','TOI-1136',0.174,0.96188,0.968,0.96188,0.968]
toig=['TOI-1136 g','TOI-1136',0.229,0.569,0.968,0.569,0.968]
mercury=['mercury','sun',0.39,0.383,1,0.383,1]
venus=['venus','sun',0.723,0.949,1,0.949,1]
mars=['mars','sun',1.52,0.532,1,0.532,1]
jupiter=['jupiter','sun',5.20,11.21,1,11.21,1]
saturn=['saturn','sun',9.57,9.45,1,9.45,1]
uranus=['uranus','sun',19.17,4.01,1,4.01,1]
neptune=['neptune','sun',30.18,3.88,1,3.88,1]

dff.loc[len(dff)] = earth
dff.loc[len(dff)]= koi
dff.loc[len(dff)]= lhs
dff.loc[len(dff)]= toib
dff.loc[len(dff)]= toic
dff.loc[len(dff)]= toid
dff.loc[len(dff)]= toie
dff.loc[len(dff)]= toif
dff.loc[len(dff)]= toig


dff = dff.reset_index(drop=True)
dff
```

```python
dff['Ds/Dp']= (dff['st_rad (Ds)']/dff['pl_rade (Dp)'])*109.7522
dff['d/Ds']=  (dff['pl_orbsmax (d)']/dff['st_rad (Ds)'])*107.1428
dff
```

```python
from matplotlib.lines import Line2D

plt.figure(figsize=(10, 6))
max_value = max(dff['st_rad (Ds)'].max(), dff['pl_orbsmax (d)'].max())

plt.plot([0, max_value], [0, max_value], color='green', linestyle='dashed', label='y = x',alpha=0.7)

# Iterate through the DataFrame and plot scatter points
for index, row in dff.iterrows():
    
    
    
    plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='blue', alpha=0.1)

    # Highlight Kepler-452 b and Kepler-62f with special markers
    if row['pl_name (Exoplanet)'] == 'Earth':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='yellow', marker='*', s=100, label='Earth')

    # Highlight Kepler-452 b and Kepler-62f with special markers
    if row['pl_name (Exoplanet)'] == 'TOI-1136 b':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='red', marker='*', s=100, label='TOI-1136 b')
    if row['pl_name (Exoplanet)'] == 'TOI-1136 c':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='purple', marker='*', s=100, label='TOI-1136 c')
    
    if row['pl_name (Exoplanet)']== 'TOI-1136 d':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='black', marker='*', s=100, label='TOI-1136 d')
    if row['pl_name (Exoplanet)']== 'TOI-1136 e':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='green', marker='*', s=100, label='TOI-1136 e')
    if row['pl_name (Exoplanet)']== 'TOI-1136 f':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='pink', marker='*', s=100, label='TOI-1136 f',alpha=0.1)
    if row['pl_name (Exoplanet)']== 'TOI-1136 g':
        plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color='orange', marker='*', s=100, label='TOI-1136 g')
    
        
# Plot the line y = x passing through the origin
max_value = max(dff['pl_rade (Dp)'].max(), dff['st_rad (Ds)'].max())
plt.plot([0, max_value], [0, max_value], color='green', linestyle='dashed', label='y = x')

# Create custom legend entries for different markers
legend_entry = Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=8, label='Exoplanets')
legend_entry_1 = Line2D([0], [0], marker='*', color='yellow', markersize=10, label='Earth')

legend_entry_3 = Line2D([0], [0], marker='*', color='red', markersize=10, label='TOI-1136 b')
legend_entry_4 = Line2D([0], [0], marker='*', color='purple', markersize=10, label='TOI-1136 c')
legend_entry_5 = Line2D([0], [0], marker='*', color='black', markersize=10, label='TOI-1136 d')
legend_entry_6 = Line2D([0], [0], marker='*', color='green', markersize=10, label='TOI-1136 e')
legend_entry_7 = Line2D([0], [0], marker='*', color='pink', markersize=10, label='TOI-1136 f')
legend_entry_8 = Line2D([0], [0], marker='*', color='orange', markersize=10, label='TOI-1136 g')





# Set the x-axis and y-axis limits
plt.xlim(0, 2)
plt.ylim(0, 2)

# Add the custom legend entries along with the line legend
plt.legend(handles=[legend_entry, legend_entry_1,legend_entry_3, legend_entry_4,legend_entry_5,legend_entry_6,legend_entry_7,legend_entry_8], loc='upper left')

# Use LaTeX-style formatting for the axis labels
plt.xlabel('Diameter of planet ($D_p$)')
plt.ylabel('Diameter of star ($D_s$)')
plt.title('Diameter of star v/s Diameter of planet \n TOI System')
plt.grid(True)
plt.show()
```
