```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('1st plot ratio-1.csv')

# Create a list with your own data for the new row
new_row_data = ['Earth', 'sun', 1,1, 1, 1, 1,109.7522, 107.1428]  # Replace 0s with your data

# Add the new row to the DataFrame
df.loc[len(df)] = new_row_data

# Reset the index
df = df.reset_index(drop=True)

# Print the DataFrame after adding the new row
df

```

```python
from matplotlib.lines import Line2D

plt.figure(figsize=(10, 6))

# Iterate through the DataFrame and plot scatter points
for index, row in df.iterrows():
    color = 'yellow' if row['pl_name (Exoplanet)'] == 'Earth' else 'blue'
    plt.scatter(row['pl_rade (Dp)'], row['st_rad (Ds)'], color=color, alpha=0.7)

# Plot the line y = x passing through the origin
max_value = max(df['pl_rade (Dp)'].max(), df['st_rad (Ds)'].max())
plt.plot([0, max_value], [0, max_value], color='green', linestyle='dashed', label='y = x')

# Create a custom legend entry for the blue data points
legend_entry = Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=8, label='Exoplanets')
legend_entry_1 = Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=8, label='Earth')
legend_entry_2 = Line2D([0], [0], linestyle='dashed', color='green', label='y = x')

# Set the x-axis and y-axis limits
plt.xlim(0, 5.22)
plt.ylim(0, 5.22)

# Add the custom legend entry along with the line legend
plt.legend(handles=[legend_entry_1, legend_entry, legend_entry_2], loc='upper left')

# Use LaTeX-style formatting for the axis labels
plt.xlabel('Diameter of planet ($D_p$)')
plt.ylabel('Diameter of star ($D_s$)')
plt.title('Plot of $D_s$ $v/s$ $D_p$')
plt.grid(True)
plt.savefig('exoplanet_plot10.pdf', format='pdf', bbox_inches='tight')
plt.show()

```
