```python
import pandas as pd
```

```python
df=pd.read_csv('Exoplanet-file1.csv')
df #original file
```

```python
kepler_452b_row = df[df['Exoplanet'] == 'Kepler-452 b']
kepler_452b_row
```

```python
df['Ds/Dp']=pd.to_numeric(df['Ds/Dp'],errors='coerce')
x=df['Ds/Dp']>=75
y=df['Ds/Dp']<=130
z1= df[(x) & (y)] #extracting ratio of Ds/Dp in range 
z1
```

```python
#extracting duplicates
import pandas as pd


# Extract the exoplanets column as a list
exoplanets_list = z1["Exoplanet"].tolist()

# Find duplicate exoplanet names
duplicate_names = set([x for x in exoplanets_list if exoplanets_list.count(x) > 1])

# Exclude rows with exoplanets having duplicate names
z1 = z1[~z1["Exoplanet"].isin(duplicate_names)]

# Save the modified DataFrame to a new CSV file
z1
```

```python
df['d/Ds']=pd.to_numeric(df['d/Ds'],errors='coerce')
x=df['d/Ds']>=75
y=df['d/Ds']<=130
z2= df[(x) & (y)]
z2
```

```python
import pandas as pd


# Extract the exoplanets column as a list
exoplanets_list = z2["Exoplanet"].tolist()

# Find duplicate exoplanet names
duplicate_names = set([x for x in exoplanets_list if exoplanets_list.count(x) > 1])

# Exclude rows with exoplanets having duplicate names
z2 = z2[~z2["Exoplanet"].isin(duplicate_names)]

# Save the modified DataFrame to a new CSV file
z2
```

```python
import pandas as pd

# Assuming you have already loaded the CSV file into the DataFrame 'df'
# df = pd.read_csv('Exoplanet-file1.csv')

# Convert 'Ds/Dp' column to numeric, ignoring non-numeric values and converting them to NaN
df.loc[:, 'Ds/Dp'] = pd.to_numeric(df['Ds/Dp'], errors='coerce')

# Convert 'd/Ds' column to numeric, ignoring non-numeric values and converting them to NaN
df.loc[:, 'd/Ds'] = pd.to_numeric(df['d/Ds'], errors='coerce')

# Combined condition
condition_combined = (df['Ds/Dp'] >= 75) & (df['Ds/Dp'] <= 130) & (df['d/Ds'] >= 75) & (df['d/Ds'] <= 130)

# Apply the combined condition to filter the DataFrame
filtered_df = df[condition_combined]
filtered_df
```
