```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# Load the CSV file
df = pd.read_csv("Ratio-2 filter-2.csv")

# Function to save dataframe as a PDF
def save_df_to_pdf(df, pdf_path):
    fig, ax = plt.subplots(figsize=(12, 8))  # Create a figure and an axes
    ax.axis('tight')
    ax.axis('off')
    table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

    pp = PdfPages(pdf_path)
    pp.savefig(fig, bbox_inches='tight')
    pp.close()

# Save the dataframe to a PDF
pdf_path = "Ratio-2_filter-2.pdf"
save_df_to_pdf(df, pdf_path)

print(f"PDF saved as {pdf_path}")
```
